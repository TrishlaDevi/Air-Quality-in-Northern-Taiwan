# Replacing the Datamarks

```python
def num(hor):
    try:
        if np.isnan(hor):
            return
        else:
            hor=str(hor)
            return float(hor.replace('x','').replace('#','').replace('*',''))
    except TypeError:
        hor=str(hor)
        return float(hor.replace('x','').replace('#','').replace('*',''))
```

# Applying the replacement function

```python
dtfr['AMB_TEMP'] = dtfr['AMB_TEMP'].apply(num)
dtfr['CH4'] = dtfr['CH4'].apply(num)
dtfr['CO'] = dtfr['CO'].apply(num)
dtfr['NMHC'] = dtfr['NMHC'].apply(num)
dtfr['NO'] = dtfr['NO'].apply(num)
dtfr['NO2'] = dtfr['NO2'].apply(num)
dtfr['NOx'] = dtfr['NOx'].apply(num)
dtfr['O3'] = dtfr['O3'].apply(num)
dtfr['PM10'] = dtfr['PM10'].apply(num)
dtfr['RH'] = dtfr['RH'].apply(num)
dtfr['SO2'] = dtfr['SO2'].apply(num)
dtfr['THC'] = dtfr['THC'].apply(num)
dtfr['UVB'] = dtfr['UVB'].apply(num)
dtfr['WD_HR'] = dtfr['WD_HR'].apply(num)
dtfr['WIND_DIREC'] = dtfr['WIND_DIREC'].apply(num)
dtfr['WIND_SPEED'] = dtfr['WIND_SPEED'].apply(num)
dtfr['WS_HR'] = dtfr['WS_HR'].apply(num)
```
